```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch, Rectangle

# 设置全局字体和样式
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 14
blue_color = '#1f77b4'  # 蓝调主色
light_blue = '#a5d8ff'  # 浅蓝背景
gray_color = 'lightgray'  # 排除盒子颜色

# 创建画布：宽12英寸，高8英寸（水平布局）
fig, ax = plt.subplots(figsize=(12, 8), dpi=300)
ax.set_title('PRISMA Flow Diagram', fontsize=14, fontweight='bold')
ax.axis('off')  # 隐藏轴

# 定义参数
box_width = 4.5  # 主盒子宽度
box_height = 1.2  # 主盒子高度
exclude_width = 3.5  # 排除盒子宽度
exclude_height = 0.8  # 排除盒子高度
x_left = -4  # 左侧阶段标签的X位置
x_main = 0  # 主流程盒子的X位置
x_exclude = 6  # 排除盒子的X位置
y_positions = [6, 4, 2, 0]  # 垂直位置（从上到下）

# 数字基于您的更新
raw_records = 639  # 假设初始	raw总记录（预去重），可调整
duplicates = 309  # 假设去重数量
total_identified = 330  # 去重后
after_screening = 303
included = 150
excluded_screening = total_identified - after_screening  # 340
excluded_fulltext = after_screening - included  # 190

# 函数：绘制圆角盒子
def draw_rounded_box(x, y, text, width=box_width, height=box_height, facecolor=light_blue, edgecolor=blue_color, rounded=True):
    boxstyle = "round,pad=0.1" if rounded else "square,pad=0.1"
    box = FancyBboxPatch((x - width/2, y - height/2), width, height, boxstyle=boxstyle, 
                         fc=facecolor, ec=edgecolor, lw=1.5)
    ax.add_patch(box)
    ax.text(x, y, text, ha='center', va='center', fontsize=14, wrap=True)

# 函数：绘制箭头
def draw_arrow(start, end, connectionstyle="arc3,rad=0", style='->'):
    arrow = FancyArrowPatch(start, end, color=blue_color, arrowstyle=style, 
                            linewidth=2, mutation_scale=20, connectionstyle=connectionstyle)
    ax.add_patch(arrow)

# 绘制左侧阶段标签（无盒子，只是文本）
phases = ['Identification', 'Screening', 'Eligibility', 'Inclusion']
for i, phase in enumerate(phases):
    ax.text(x_left, y_positions[i], phase, ha='right', va='center', fontsize=14, fontweight='bold', color=blue_color)

# 1. Identification Phase
id_text = (f'\nScopus, WoS , Google Scholar \n'
           f'Initial raw total (n={raw_records})\nDuplicates removed (n={duplicates})\nTotal after deduplication (n={total_identified})')
draw_rounded_box(x_main, y_positions[0], id_text)

# Arrow down from Identification to Screening
draw_arrow((x_main, y_positions[0] - 0.7), (x_main, y_positions[1] + 0.7))

# 2. Screening Phase
screen_text = f'Records screened \nby title and abstract\n(n={total_identified})'
draw_rounded_box(x_main, y_positions[1], screen_text)

# Side exclusion for screening
draw_arrow((x_main + box_width/2, y_positions[1]), (x_exclude - exclude_width/2, y_positions[1]), connectionstyle="arc3,rad=0")
exclude_text = f'Records excluded\n(n={excluded_screening})'
draw_rounded_box(x_exclude, y_positions[1], exclude_text, width=exclude_width, height=exclude_height, facecolor=gray_color, edgecolor='gray', rounded=False)

# Arrow down from Screening to Eligibility
draw_arrow((x_main, y_positions[1] - 0.7), (x_main, y_positions[2] + 0.7))

# 3. Eligibility Phase
elig_text = f'Full-text articles \nassessed for eligibility\n(n={after_screening})'
draw_rounded_box(x_main, y_positions[2], elig_text)

# Side exclusion for full-text
draw_arrow((x_main + box_width/2, y_positions[2]), (x_exclude - exclude_width/2, y_positions[2]), connectionstyle="arc3,rad=0")
exclude_text_full = (f'Full-text articles excluded\n(n={excluded_fulltext})\n'
                     f'• No clear motivation (n=80)\n• Focus out of scope (n=73)')
draw_rounded_box(x_exclude, y_positions[2], exclude_text_full, width=exclude_width, height=exclude_height + 0.4, facecolor=gray_color, edgecolor='gray', rounded=False)

# Arrow down from Eligibility to Inclusion
draw_arrow((x_main, y_positions[2] - 0.7), (x_main, y_positions[3] + 0.7))

# 4. Inclusion Phase
incl_text = f'Studies included \nin the systematic review\n(n={included})'
draw_rounded_box(x_main, y_positions[3], incl_text)

# 绘制外边大方框包围整个图
outer_box = Rectangle((-6, -1), 14, 8.5, fill=False, edgecolor=blue_color, linewidth=2, linestyle='--')
ax.add_patch(outer_box)

# 调整布局并保存为PNG
plt.xlim(-6, 10)
plt.ylim(-1.5, 7.5)
plt.tight_layout()
plt.savefig('3.prisma_output.pdf', format='pdf', bbox_inches='tight', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle, FancyArrowPatch

# 设置全局字体
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 10
light_red = '#ff6666'  # 淡红色 for RSSI
light_blue_color = '#6666ff'  # 淡蓝色 for CSI
background_blue = '#a5d8ff'  # 浅蓝背景

# 函数：生成 Figure 3.1 (RSSI vs CSI Comparison) - 淡化颜色，文本上移
def generate_rssi_csi_figure():
    # 模拟数据：时间序列 (0-10秒)，基于真实HAR模式 (e.g., 行走引起的周期性+多径噪声)
    time = np.linspace(0, 10, 500)  # 时间轴
    # RSSI: 粗粒度，聚合噪声 (基于 \citep{guo2019robust} 的典型模式)
    rssi = 0.5 + 0.2 * np.sin(2 * np.pi * 0.5 * time) + 0.15 * np.random.normal(size=len(time))  # 高噪声
    # CSI: 细粒度，多径效应 (添加多个频率分量模拟子载波, per \citep{venkatnarayan2018multiperson})
    csi = 0.5 + 0.3 * np.sin(2 * np.pi * 0.5 * time + np.pi/4) + 0.05 * np.random.normal(size=len(time)) + 0.1 * np.sin(2 * np.pi * 2 * time) + 0.08 * np.sin(2 * np.pi * 3 * time)  # 多径细节

    # 归一化
    rssi = (rssi - np.min(rssi)) / (np.max(rssi) - np.min(rssi))
    csi = (csi - np.min(csi)) / (np.max(csi) - np.min(csi))

    # 绘制
    fig, ax = plt.subplots(figsize=(6, 4), dpi=300)
    ax.plot(time, rssi, label='RSSI (Coarse-Grained)', color=light_red, linewidth=2, linestyle='--')
    ax.plot(time, csi, label='CSI (Fine-Grained)', color=light_blue_color, linewidth=2)
    ax.set_title('RSSI vs CSI During Human Walking', fontsize=12)
    ax.set_xlabel('Time (seconds)')
    ax.set_ylabel('Normalized Signal Strength')
    ax.set_ylim(0, 1.1)
    ax.legend(loc='upper right')
    ax.grid(True, linestyle='--', alpha=0.5)
    # 文本上移，避免重叠
    ax.text(4, 1.0, 'CSI captures multipath distortions\n(Up to 25% higher accuracy)', ha='center', fontsize=9, color=light_blue_color)

    plt.tight_layout()
    plt.savefig('4.rssi_csi_figure.png', format='png', bbox_inches='tight')
    plt.show()

# 函数：生成 Figure 3.2 (CSI Acquisition Workflow) - 垂直排列，确保字体清晰
def generate_csi_workflow_figure():
    # 创建画布 (垂直布局)
    fig, ax = plt.subplots(figsize=(6, 6), dpi=300)  # 增加高度以适应垂直排列
    ax.set_title('CSI Acquisition Workflow with Protocol Evolution', fontsize=12)
    ax.axis('off')

    # 步骤节点 (垂直排列，体现流程：传输→获取→处理→应用)
    nodes = [
        'Signal Transmission\n(Packets via Wi-Fi)\n802.11n: Basic MIMO',
        'CSI Acquisition\n(Capture & Parse Subcarriers)\nWi-Fi 6: MU-MIMO Enhances',
        'Signal Processing\n(Denoising & Calibration)\nWi-Fi 7: Sensing Std Improves',
        'HAR Application\n(Activity Classification)\nReal-Time Output'
    ]
    y_positions = [0.85, 0.6, 0.35, 0.1]  # 垂直位置

    rect_width = 0.8  # 增加宽度以适应文本
    rect_height = 0.15  # 增加高度以提高清晰度
    x_center = 0.5

    for i, node in enumerate(nodes):
        y_center = y_positions[i]
        llx = x_center - rect_width / 2
        lly = y_center - rect_height / 2
        ax.add_patch(Rectangle((llx, lly), rect_width, rect_height, fill=True, facecolor=background_blue, edgecolor=light_blue_color, linewidth=1.5))
        ax.text(x_center, y_center, node, va='center', ha='center', fontsize=9, color='black', wrap=True)  # 增大字体以清晰

    # 添加箭头 (垂直方向)
    for i in range(len(nodes) - 1):
        arrow_start_y = y_positions[i] - rect_height / 2 - 0.01
        arrow_end_y = y_positions[i + 1] + rect_height / 2 + 0.01
        ax.add_patch(FancyArrowPatch((0.5, arrow_start_y), (0.5, arrow_end_y), color=light_blue_color, arrowstyle='->', linewidth=2, mutation_scale=20, connectionstyle="arc3,rad=0"))

    # 外边框
    ax.add_patch(Rectangle((0.05, 0), 0.9, 0.95, fill=False, edgecolor=light_blue_color, linewidth=2))

    plt.tight_layout()
    plt.savefig('5.CSI workflow.png', format='png', bbox_inches='tight')
    plt.show()

# 运行生成函数
generate_rssi_csi_figure()
generate_csi_workflow_figure()
```

```python
import plotly.graph_objects as go
import numpy as np

# 定义神经网络层（基于Hybrid模型：input=15, hidden=128, output=5）
layers = [
    {'name': 'Input (Hybrid Features)', 'nodes': 15, 'color': 'blue'},  # 15 = PCA 10 + Physics 5
    {'name': 'LSTM Hidden', 'nodes': 128, 'color': 'green'},
    {'name': 'Output', 'nodes': 5, 'color': 'red'}
]

# 生成3D坐标：X=层索引，Y/Z=随机分布（模拟节点位置）
node_positions = []
connections = []
current_x = 0
for i, layer in enumerate(layers):
    y = np.random.uniform(-1, 1, layer['nodes'])
    z = np.random.uniform(-1, 1, layer['nodes'])
    x = np.full(layer['nodes'], current_x)
    node_positions.append(go.Scatter3d(
        x=x, y=y, z=z, 
        mode='markers+text',
        marker=dict(size=5, color=layer['color']),
        #text=[f"{layer['name']} Node {j}" for j in range(layer['nodes'])],
        name=layer['name']
    ))
    
    # 连接到下一层（全连接模拟）
    if i < len(layers) - 1:
        next_layer_nodes = layers[i+1]['nodes']  # 修正：直接获取下一层节点数（整数）
        next_y = np.random.uniform(-1, 1, next_layer_nodes)  # 预生下一层Y坐标
        next_z = np.random.uniform(-1, 1, next_layer_nodes)  # 预生下一层Z坐标
        for src in range(layer['nodes']):
            for tgt in range(next_layer_nodes):
                connections.append(go.Scatter3d(
                    x=[x[src], current_x + 1],
                    y=[y[src], next_y[tgt]],  # 使用预生成的坐标，避免重复随机
                    z=[z[src], next_z[tgt]],
                    mode='lines', 
                    line=dict(color='gray', width=1),
                    showlegend=False  # 线条不显示在图例中
                ))
    current_x += 1

# 创建3D图
fig = go.Figure(data=node_positions + connections)
fig.update_layout(
    title='3D Neural Network Structure (Hybrid LSTM Model)',
    scene=dict(
        xaxis_title='Layer', 
        yaxis_title='Y', 
        zaxis_title='Z',
        aspectmode='cube'
    ),
    width=800, 
    height=600
)

# 保存为交互HTML和静态PNG
fig.write_html('nn_3d1.html')
#fig.write_image('nn_3d.png')
fig.write_image('nn_3d1.png', engine='orca')

print("3D图已保存为 'nn_3d.html' (交互) 和 'nn_3d.png' (静态)")
```

```python
import torch
import torch.nn as nn
from torchviz import make_dot

# 定义Hybrid模型（LSTM部分）
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first=True)
        self.dropout = nn.Dropout(0.2)
        self.fc = nn.Linear(hidden_size, num_classes)
    
    def forward(self, x):
        out, _ = self.lstm(x)
        out = self.dropout(out[:, -1, :])
        out = self.fc(out)
        return out

# 实例化模型
model = LSTMModel(input_size=15, hidden_size=128, num_classes=5)
x = torch.randn(1, 100, 15)  # 虚拟输入 (batch=1, seq=100, features=15)

# 生成图
y = model(x)
dot = make_dot(y, params=dict(model.named_parameters()), show_attrs=True, show_saved=True)
dot.render('hybrid_torchviz', format='png')  # 或 'pdf'
print("图已保存为 'hybrid_torchviz.png'")
```
